```python
import options_risk as e
import numpy as np  
```

```python
# Option call européenne avec volatilité élevée et long maturity
params_eu = {
    'S0': 100,       # Prix initial
    'K': 100,        # Prix d'exercice
    'T': 2.0,        # Temps jusqu'à maturité (2 ans)
    'r': 0.05,       # Taux d'intérêt sans risque (5%)
    'sigma': 0.40,   # Volatilité élevée (40%)
    'option_type': 'call',
    'position': 'long',
    'type_option': 'eu',
    'dividend_yield': 0.03  # Taux de dividende (3%)
}
```

```python
# Créer l'instance d'option avec les paramètres souhaités
option = e.EuropeanOptionMonteCarlo(**params_eu)  # ou tout autre ensemble de paramètres

# Exécuter la simulation
results = option.simulate_option_price(num_simulations=10000, steps=252)

# Afficher les résultats et la différence avec BS
print(f"Prix Monte Carlo: {option.option_price:.4f}")
print(f"Prix Black-Scholes: {option.bs_price:.4f}")
print(f"Différence absolue: {abs(option.option_price - option.bs_price):.4f}")
print(f"Différence relative: {100 * abs(option.option_price - option.bs_price) / option.bs_price:.2f}%")

# Visualiser la distribution des prix
option.plot_price_distribution()

# Visualiser les métriques de risque
option.plot_risk_metrics()
```
